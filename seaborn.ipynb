```python
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("T:/project/data_science_lab/titanic.csv")
```

```python
df.head()
```

```python
plt.figure(figsize=(5,3))
sns.scatterplot(data=df, x='Age',y='Fare')
plt.title("Scatter Plot: Age vs Fare by Survival")
plt.show()
```

```python
plt.figure(figsize=(5,3))
sns.countplot(data=df, x='Sex')
plt.title("Survival by gender")
plt.show()
```

```python
plt.figure(figsize=(5,3))
sns.barplot(data=df, x='Pclass', y='Survived')
plt.title("Survival by gender")
plt.show()
```

```python
plt.figure(figsize=(5,3))
sns.boxplot(data=df, x='Pclass', y='Fare')
plt.title("Survival by gender")
plt.show()
```

```python
plt.figure(figsize=(5,3))
sns.violinplot(data=df, x='Survived', y='Age')
plt.title("Survival by gender")
plt.show()
```

```python
plt.figure(figsize=(5,3))
sns.kdeplot(data=df[df['Survived']==0], x='Fare')
sns.kdeplot(data=df[df['Survived']==1], x='Fare')
plt.title("Survival by gender")
plt.show()
```

```python
data = df.select_dtypes(include=['float64','int64'])
corr = data.corr()
plt.figure(figsize=(5,3))
sns.heatmap(corr)
plt.title("Survival by gender")
plt.show()
```
